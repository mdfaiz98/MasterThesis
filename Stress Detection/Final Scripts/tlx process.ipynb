```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.typing import ArrayLike
import seaborn as sns
```

```python
TLX = pd.read_csv("C:/Thesis-script/Data/Nasa-Tlx-zscores.csv")
TLX.rename(columns={'ParticipantID': 'Participant ID'}, inplace=True)
count_weighted_rating_gt_50 = (TLX['WeightedRating'] > 50).sum()

plt.figure(figsize=(10, 6))
sns.histplot(TLX['Z_Score'], bins=30, kde=True)
plt.title('Distribution of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.axvline(x=0, color='red', linestyle='--', label='Lower Threshold (-0.5)')
plt.axvline(x=1, color='green', linestyle='--', label='Upper Threshold (0.5)')
plt.legend()
plt.show()

```
